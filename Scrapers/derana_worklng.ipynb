```python
import pandas as pd
from bs4 import BeautifulSoup
from lxml import html
import requests
import re
import time
```

```python
import nltk
nltk.download('vader_lexicon')

from nltk.sentiment.vader import SentimentIntensityAnalyzer

sid = SentimentIntensityAnalyzer()
```

## Phase 01: Grabbing all URLs to be scraped

```python
test_query = 'canada'
```

```python
base_string = 'http://www.adaderana.lk/search_results.php?mode=2&show=1&query='
```

```python
derana_base = 'http://www.adaderana.lk/'
```

```python
URL = base_string + test_query 
```

```python
headers = {'User-Agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10.15; rv:98.0) Gecko/20100101 Firefox/98.0'}
```

```python
page = requests.get(URL, headers = headers)
```

```python
soup = BeautifulSoup(page.content, "html.parser")
```

```python
news_cards = soup.find_all(
            "div", attrs={"class": "news-story"})
```

```python
len(news_cards)
```

```python
temp_list = []

for data in news_cards:
    
    url = data.find_all(href = True)
    
    temp_list.append(url)
```

```python
result_list = []

for i in range(len(temp_list)):
   
    a_tag = str(temp_list[i][0])
    
    clean_tag = a_tag[0:51]
    clean_tag = re.search(r'href=(.*)', clean_tag, re.DOTALL)
    clean_tag = clean_tag.group(1)
    clean_tag = clean_tag.replace('"',"")
    final_url = clean_tag.replace(">","")
    
    result_list.append(final_url)
    
    
```

```python
final_list = []

for item in result_list:
    url = derana_base + item
    final_list.append(url)
    
```

```python
len(final_list)
```

```python
final_list[0]
```

## Phase 2: Scraping each article page

```python
database = []
i = 0

for url in final_list:
  
    record = []
    print(url)
    
    page = requests.get(url, headers = headers)
    soup = BeautifulSoup(page.content, "html.parser")
    
    try:
        name = soup.find("h1").text
        date = soup.find("p",attrs={"class":"news-datestamp"}).text.strip()
        body = soup.find("div",attrs={"class":"news-content"}).text.strip()
    except:
        print('An element was not found')
    
    record.append(name)
    record.append(date)
    record.append(body)
    
    database.append(record)
    i += 1
    print('Record ' + str(i) + ' complete ...')
    
    time.sleep(5)
```

```python
df = pd.DataFrame(database, columns = ['ArticleTitle', 'Date','Body'])
```

```python
df.head()
```

## Phase 3: VADER for Sentiment Analysis

```python
df['scores'] = df['Body'].apply(lambda body: sid.polarity_scores(body))
```

```python
df.head()
```

```python
df['compound']  = df['scores'].apply(lambda score_dict: score_dict['compound'])
```

```python
df['news_grade'] = df['compound'].apply(lambda c: 'pos' if c >=0.9 else 'neg' if c <= -0.9 else 'neutral')
```

```python
df.to_csv('Derana_Results.csv',index=False)
```
